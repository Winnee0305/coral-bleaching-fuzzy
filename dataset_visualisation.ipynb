### Cleaning dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('./global_bleaching_environmental.csv')
print(df.shape)       # rows and columns
print(df.info())      # data types and null counts
print(df.head())      # preview first few rows
print(df.describe())  # summary of numerical columns
```

```python
# Drop rows with any missing values
df = df.dropna()

# Or drop columns with too many missing values
df = df.dropna(axis=1, thresh=len(df)*0.5)  # drop columns with >50% missing
```

```python
# Find duplicates
duplicates = df[df.duplicated()]

# Drop duplicates
df = df.drop_duplicates()
```

```python
# Final sanity checks
print(df.info())
print(df.isna().sum())

# Save cleaned data
df.to_csv('cleaned_data.csv', index=False)
```
