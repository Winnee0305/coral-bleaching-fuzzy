```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./cleaned_data.csv')

df.head()

```

```python
def print_mix_max(data, label):
    data = pd.to_numeric(data, errors='coerce')
    print(f"{label} - Min: {data.min()}, Max: {data.max()}")
```

```python
def print_corr(var1, var2, label):    
    var1 = pd.to_numeric(var1, errors='coerce')
    var2 = pd.to_numeric(var2, errors='coerce')
    corr = var1.corr(var2)
    print(f"{label} - Correlation: {corr}")
    return corr
```

```python
def plot_scatter_plot(y, x, y_label, x_label, title):
    x = pd.to_numeric(x, errors='coerce')
    y = pd.to_numeric(y, errors='coerce')

    plt.figure(figsize=(8, 6))
    plt.scatter(x, y)
    
    # Titles and labels
    plt.title(title, fontsize=14)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
def plot_histogram(data, label, title, bins=30):
    data = pd.to_numeric(data, errors='coerce')

    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=bins, edgecolor='black', alpha=0.7)
    
    # Titles and labels
    plt.title(title, fontsize=14)
    plt.xlabel(label, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

### Sea Water Temperature

```python
print_mix_max(df['Temperature_Mean'], 'Temperature Mean')
print_corr(df['Temperature_Mean'], df['Percent_Bleaching'], 'Temperature Mean vs. Percent Bleaching')
plot_scatter_plot(df['Temperature_Mean'], df['Percent_Bleaching'], 'Temperature Mean', 'Percent Bleaching', 'Temperature Mean vs. Percent Bleaching')
plot_histogram(df['Temperature_Mean'], 'Temperature Mean', 'Histogram of Temperature Mean')
```

```python
print_mix_max(df['ClimSST'], 'ClimSST')
print_corr(df['ClimSST'], df['Percent_Bleaching'], 'ClimSST vs. Percent Bleaching')
plot_scatter_plot(df['ClimSST'], df['Percent_Bleaching'], 'ClimSST', 'Percent Bleaching', 'ClimSST vs. Percent Bleaching')
plot_histogram(df['ClimSST'], 'ClimSST', 'Histogram of ClimSST')
```
